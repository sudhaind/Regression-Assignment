```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
dependent=dataset[["charges"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Creating the model here
#The model is saved in regressor
from sklearn.tree import DecisionTreeRegressor
regressor =DecisionTreeRegressor(criterion='squared_error',splitter='random',max_features='sqrt')
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```
